```python
# Total revenue by top 20 model -- Bar chart
df3 = df.groupby('Model')['Revenue_USD'].sum().head(20).sort_values(ascending=False)
plt.bar(df3.index,df3.values,color='g')
plt.xlabel('Model')
plt.ylabel('Revenue total')
plt.title('TOTAL REVENUE BY MODEL')
plt.xticks(rotation=90)
```

```python
# Total price USD by top 10 brand -- horizontal bar chart
df4 = df.groupby('Brand')['Price_USD'].sum().head(10).sort_values(ascending=False)
plt.barh(df4.index,df4.values,color='b')
plt.xlabel('Brand')
plt.ylabel('price')
plt.title('TOTAL PRICE BY BRAND')
```

```python
# distribution of price USD
plt.hist(df['Price_USD'],bins=10,color='r')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('PRICE DISTRIBUTION')
```

```python
# Units sold vs Sales month -- scatter plot
plt.scatter(df['Units_Sold'],df['Sale_Month'])
plt.xlabel('units sold')
plt.ylabel('sales month')
plt.title('UNITS SOLD vs SALES MONTH')
```

```python
# which brand has highest unit sold above 15 -- Count plot
df_15 = df[df['Units_Sold']>=15]
lb=sns.countplot(data=df_15,x='Brand')
plt.xticks(rotation=90)
for vl in lb.containers:
  lb.bar_label(vl)
```

```python
plt.figure(figsize=(10,8))
sns.histplot(data=df,x='Units_Sold',bins=10,hue='Country',multiple='stack')
```
